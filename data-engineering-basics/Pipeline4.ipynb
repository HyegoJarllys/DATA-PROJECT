```python
api_data = [
{"city":"São Paulo","temperature":28,"humidity":65},
{"city":"Rio de Janeiro","temperature":32,"humidity":70},
{"city":"Belo Horizonte","temperature":26,"humidity":60},
{"city":"Curitiba","temperature":22,"humidity":75},
{"city":"Porto Alegre","temperature":24,"humidity":68}
]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar DataFrame
df=pd.DataFrame(api_data)
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
#metricas
temperatura_media= df["temperature"].mean()
print(f"A temperatura média é: {temperatura_media:.2f}°C")
cidade_mais_quente= df.groupby("city")["temperature"].max().nlargest(1)
print(f"A cidade mais quente é: {cidade_mais_quente.index[0]}")
```

```python
#gráfico de temperatura por cidade
plt.figure(figsize=(8,6))
plt.bar(df["city"], df["temperature"], color="orange")
plt.title("Temperatura por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Temperatura (°C)")
plt.xticks(rotation=45)
plt.show()
```

```python
#grafico de umidade por cidade
plt.figure(figsize=(8,6))
plt.plot(df["city"],df["humidity"],color="blue",marker="o")
plt.title("Umidade por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Umidade (%)")
plt.xticks(rotation=45)
plt.show()
```
