```python
🎯 Objetivo

Transformar API de pedidos em um dataframe e extrair insights de vendas.

🟡 Tarefa 1 — Criar dataframe da API
🟡 Tarefa 2 — Explodir a coluna items
🟡 Tarefa 3 — Normalizar os produtos
🟡 Tarefa 4 — Criar coluna total_sale
🟡 Tarefa 5 — Métricas
Calcular:

1️⃣ faturamento total
2️⃣ produto mais vendido
3️⃣ cliente que mais gastou
```

```python
import pandas as pd
import numpy as np
```

```python
api_data = [
 {
  "order_id":1,
  "customer":"Ana",
  "city":"SP",
  "items":[
      {"product":"Notebook","price":3500,"quantity":1},
      {"product":"Mouse","price":100,"quantity":2}
  ]
 },
 {
  "order_id":2,
  "customer":"João",
  "city":"RJ",
  "items":[
      {"product":"Monitor","price":1200,"quantity":1}
  ]
 }
]
```

```python
#1
df= pd.DataFrame(api_data)
print(df)
```

```python
#2
df= df.explode("items") 
print(df)
```

```python
#3
items= pd.json_normalize(df["items"]) 
print(items) 
```

```python
df=df.drop(columns="items") 
print(df) 
```

```python
df= pd.concat([df, items], axis=1)
print(df)
```

```python
#4
df["total_sale"]= df["price"]*df["quantity"] 
print(df)
```

```python
#faturamento total
tot_fat= df["total_sale"].sum() 
print(f"O faturamento total é: R${tot_fat:.2f}") 
```

```python
#mais vendido
mais_vendido= df.groupby("product")["quantity"].sum().nlargest(1) 
print(f"o produto mais vendido é: {mais_vendido.index[0]}") 
```

```python
#melhor cliente
cliente_mais_gasto= df.groupby("customer")["total_sale"].sum().nlargest(1) 
print(f"O cliente que mais gastou é: {cliente_mais_gasto.index[0]}")
```
