```python

Imagine que você recebeu 3 datasets do time de engenharia.

Seu objetivo é criar métricas de negócio.

🎯 Objetivo

Construir uma tabela final de vendas.


🟡 Tarefa 1 — Criar DataFrames

🟡 Tarefa 2 — Primeiro merge

🟡 Tarefa 3 — Segundo merge

🟡 Tarefa 4 — Criar coluna total_sale

🟡 Tarefa 5 — Faturamento total

🟡 Tarefa 6 — Produto mais vendido

🟡 Tarefa 7 — Top 3 clientes que mais gastaram

🟡 Tarefa 8 — Faturamento por mês
```

```python
#📦 Dataset 1 — Orders
orders = [
 {"order_id":1,"customer":"Ana","date":"2024-01-10"},
 {"order_id":2,"customer":"João","date":"2024-01-11"},
 {"order_id":3,"customer":"Maria","date":"2024-01-12"},
 {"order_id":4,"customer":"Ana","date":"2024-02-01"},
]

#📦 Dataset 2 — Order Items
order_items = [
 {"order_id":1,"product_id":101,"quantity":1},
 {"order_id":1,"product_id":102,"quantity":2},
 {"order_id":2,"product_id":103,"quantity":1},
 {"order_id":3,"product_id":104,"quantity":1},
 {"order_id":4,"product_id":101,"quantity":1},
]


#📦 Dataset 3 — Products
products = [
 {"product_id":101,"product":"Notebook","price":3500},
 {"product_id":102,"product":"Mouse","price":100},
 {"product_id":103,"product":"Monitor","price":1200},
 {"product_id":104,"product":"Teclado","price":200},
]
```

```python
#1
import pandas as pd
df_orders= pd.DataFrame(orders) 
df_items=pd.DataFrame(order_items) 
df_products=pd.DataFrame(products) 
print(df_orders) 
print (df_items) 
print (df_products)
```

```python
#2
primeiro_merge= pd.merge(df_orders, df_items, on="order_id", how="inner") 
print(primeiro_merge)
```

```python
#3
df_final=pd.merge(primeiro_merge, df_products, on="product_id", how="inner") 
print(df_final)
```

```python
#4
df_final["total_sale"] = df_final["quantity"]*df_final["price"] 
print(df_final)
```

```python
#5
faturamento_total= df_final["total_sale"].sum() 
print(f"O faturamento total é: R${faturamento_total:.2f}")
```

```python
#6
product_sale= df_final.groupby("product")["quantity"].sum().nlargest(1) 
print(f"O produto mais vendido é: {product_sale}")
```

```python
#7
melhores_clients= df_final.groupby("customer")["total_sale"].sum().nlargest(3) 
print(f"Os clientes que mais gastaram são: {melhores_clients}")
```

```python
#8
df_final["date"]=pd.to_datetime(df_final["date"]) 
df_final.info() 
```

```python
df_final["mes"]= df_final["date"].dt.month 
print(df_final) 
```

```python
faturamento_mensal=df_final.groupby("mes")["total_sale"].sum() 
print(faturamento_mensal)
```
