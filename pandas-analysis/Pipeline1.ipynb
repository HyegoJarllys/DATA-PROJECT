```python
import pandas as pd
import numpy as np
```

```python
🟢 Mini Projeto 1 — Pipeline de Análise de Vendas
```

```python
#dataset


data = [
 {"order_id":1,"customer":"Ana","city":"SP","product":"Notebook","price":3500,"quantity":1},
 {"order_id":2,"customer":"João","city":"RJ","product":"Mouse","price":100,"quantity":2},
 {"order_id":3,"customer":"Ana","city":"SP","product":"Teclado","price":200,"quantity":1},
 {"order_id":4,"customer":"Maria","city":"MG","product":"Monitor","price":1200,"quantity":1},
 {"order_id":5,"customer":"João","city":"RJ","product":"Notebook","price":3500,"quantity":1},
 {"order_id":6,"customer":"Carlos","city":"SP","product":"Mouse","price":100,"quantity":3},
]
```

```python
🎯 Objetivo do pipeline

Gerar métricas de negócio.

🟡 Tarefa 1 — Criar coluna total_sale


🟡 Tarefa 2 — Top clientes


🟡 Tarefa 3 — Vendas por cidade


🟡 Tarefa 4 — Produto mais vendido (quantidade)


🟡 Tarefa 5 — Ticket médio por pedido


🟡 Tarefa 6 — Cidade com maior faturamento
```

```python
#1-
df=pd.DataFrame(data)
df["total_sale"]=df["price"]*df["quantity"] 
print(df)
```

```python
#2
gasto_cliente = df.groupby("customer")["total_sale"].sum()
```

```python
#3
vendas_cidade = df.groupby("city")["total_sale"].sum()
```

```python
#4
produtos_mais_vendidos = df.groupby("product")["quantity"].count()
```

```python
#5
ticket_medio = df["total_sale"].mean()
```

```python
#6
maior_faturamento = df.groupby("city")["total_sale"].sum().nlargest(1)
```

```python
#7
report={
 "gasto_cliente": gasto_cliente,
 "vendas_cidade": vendas_cidade,
 "produtos_mais_vendidos": produtos_mais_vendidos,
 "ticket_medio": ticket_medio,
 "maior_faturamento": maior_faturamento
}
print(report)
```
