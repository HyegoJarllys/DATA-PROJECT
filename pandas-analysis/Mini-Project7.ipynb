```python
#criação do Data Frame
caminho_origem=r"C:\Users\Hyego Jarllys\.cache\kagglehub\datasets\vivek468\superstore-dataset-final\versions\1\Sample - Superstore.csv"
df = pd.read_csv(caminho_origem, encoding='windows-1252')
```

```python
#TRANSFORMAÇÂO
df.shape
df.head()
```

```python
df=df.drop_duplicates()
df.isnull().sum()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

```python
#Gravar no DataBase

import sqlite3

conn = sqlite3.connect("Superstore.db")

df.to_sql("Superstore", conn, if_exists="replace", index=False)
df.head(1)
```

```python
#Queryes
query=  """
select "Product Name", Quantity as 'Quantidade'
from Superstore
order by Quantity desc
limit 1
"""
mais_vendido=pd.read_sql(query,conn)
print(mais_vendido)
```

```python
query="""
select Category, round(sum(Profit),2) as ' Receita'
from Superstore
group by Category
order by sum(Profit) desc
"""
maior_categoria=pd.read_sql(query,conn)
print(maior_categoria)
```

```python
query = """
SELECT strftime('%m/%Y', [Order Date]) as Mes_Ano,
       ROUND(SUM(Sales), 2) as Faturamento
FROM Superstore
GROUP BY Mes_Ano
ORDER BY Faturamento DESC
LIMIT 10
"""

df_mensal = pd.read_sql(query, conn)
print(df_mensal)
```

```python
'''
1️⃣ Qual produto vende mais?
2️⃣ Qual categoria gera mais receita?
3️⃣ Qual mês teve maior faturamento?
4️⃣ Qual ticket médio por pedido?
# 1. A Pergunta (Query)
# Escrevemos em SQL puro dentro de uma variável de texto (string)
query = """
SELECT Category, SUM(Sales) as Total_Vendas
FROM sales
GROUP BY Category
ORDER BY Total_Vendas DESC
"""

# 2. O Mensageiro (Pandas + Connection)
# O pd.read_sql pega a query e a conexão 'conn' para buscar o dado
df_resultado = pd.read_sql(query, conn)

# 3. A Resposta
# O resultado volta para você já formatado como um novo DataFrame do Pandas
print(df_resultado)
'''
```
