```python
#criação do Data Frame
caminho_origem=r"C:\Users\Hyego Jarllys\.cache\kagglehub\datasets\vivek468\superstore-dataset-final\versions\1\Sample - Superstore.csv"
df = pd.read_csv(caminho_origem, encoding='windows-1252')
```

```python
#TRANSFORMAÇÂO
df.shape
df.head()
```

```python
df=df.drop_duplicates()
df.isnull().sum()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

```python
#Gravar no DataBase

import sqlite3

conn = sqlite3.connect("Superstore.db")

df.to_sql("Superstore", conn, if_exists="replace", index=False)
df.head(1)
```

```python
#Queryes
query=  """
select "Product Name", Quantity as 'Quantidade'
from Superstore
order by Quantity desc
limit 10
"""
mais_vendido=pd.read_sql(query,conn)
print(mais_vendido)
```

```python
query="""
select Category, round(sum(Profit),2) as ' Receita'
from Superstore
group by Category
order by sum(Profit) desc
"""
maior_categoria=pd.read_sql(query,conn)
print(maior_categoria)
```

```python
query = """
SELECT strftime('%m/%Y', [Order Date]) as Mes_Ano,
       ROUND(SUM(Sales*Quantity), 2) as Faturamento
FROM Superstore
GROUP BY Mes_Ano
ORDER BY strftime('%Y', [Order Date]), strftime('%m', [Order Date])
"""

df_mensal = pd.read_sql(query, conn)
print(df_mensal)
```

```python
query="""
select sum(Sales*Quantity)/ count(DISTINCT"Order ID") as 'ticket medio'
from Superstore
"""
ticket_medio= pd.read_sql(query,conn)
print(ticket_medio)
```
